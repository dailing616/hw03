```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#import data
data = pd.read_csv('Germany_Renewable_Energy.csv')
data.head()
```

```python
#visualization

plt.figure(figsize=(20,15))

ax1 = plt.subplot2grid((2, 4), (0, 0), colspan=2, rowspan=2)
ax2 = plt.subplot2grid((2, 4), (0, 2), colspan=2)
ax3 = plt.subplot2grid((2, 4), (1, 2), colspan=2)

x = data['Year']
y = [data['Hydro'],data['Biogenic waste'],data['Biomass'],
     data['Wind (onshore)'],data['Wind (offshore)'],data['PV']]

# Basic stacked area chart.
ax1.stackplot(x,y, labels=['Hydroelectricity','Waste Incernation','Biomass',
                           'Wind (onshore)', 'Wind (offshore)','Solar Power (PV)'],
              colors=sns.xkcd_palette(["sea blue",'coral',"silver",
                                       "faded green",'baby blue',"amber"]),
              alpha=0.75)
ax1.legend(loc='upper left', prop={'size': 12})
ax1.set_title('Electric Gross Generation in Germany by Renewable Sources Since 1990',
               y=1.022, fontsize=16)
ax1.set_xlabel('Year', fontsize=14)
ax1.set_ylabel('Electric Gross Generation (GW⋅h)', fontsize=14)

#subplot2
ax2.set_ylim(0,100)
ax2.stackplot(x,data['% of Total'],(100-data['% of Total']),
              labels = ['Renewable','Non-Renewable'], alpha = 0.4)
ax2.set_title('Renewable Energy as a % of Total Electric Gross Generation',
               y=1.05, fontsize=16)
ax2.set_xlabel('Year', fontsize=14)
ax2.set_ylabel('% of Total Electric Gross Generation', fontsize=14)
ax2.legend(loc='upper left', prop={'size': 12})
ax2.yaxis.tick_right()
ax2.yaxis.set_label_position('right')

#subplot3
ax3.set_title('Electric Gross Generation in Germany by Different Renewable Sources',
               y=1.05, fontsize=16)
ax3.plot(data['Year'],data['PV'],color=sns.xkcd_palette(["amber"])[0])
ax3.plot(data['Year'],data['Wind (onshore)'],color=sns.xkcd_palette(["faded green"])[0])
ax3.plot(data['Year'],data['Wind (offshore)'],color=sns.xkcd_palette(["baby blue"])[0])
ax3.plot(data['Year'],data['Biomass'],color=sns.xkcd_palette(["silver"])[0])
ax3.plot(data['Year'],data['Hydro'],color=sns.xkcd_palette(["sea blue"])[0])
ax3.plot(data['Year'],data['Biogenic waste'],color=sns.xkcd_palette(["coral"])[0])
ax3.legend(loc='upper left', prop={'size': 12})
ax3.set_xlabel('Year', fontsize=14)
ax3.set_ylabel('Electric Gross Generation (GW⋅h)', fontsize=14)
ax3.yaxis.tick_right()
ax3.yaxis.set_label_position('right')

plt.suptitle('Renewable Energy in Germany', fontsize=24)

plt.show()
```

I designed a relatively simple visualization above to demonstrate the change of electricity production in Germany, "the world's first major renewable energy economy". The visualization consists of three subplots: a stacked area plot (left), a 100% stacked area plot (upper-right), and a line plot (lower-left).

There are several stories that I wanted to tell through this visualization. First, I wanted to show how the electric gross generation by renewable energy courses has changed in Germany since 1990, both in terms of total amount and in terms of composition. According to the first subplot (left), we can clearly see that the electric gross generation by renewable energy sources has increased by around 10 times. Second, I also wanted to show how the electric gross generation by renewable energy sources has changed as a percentage of the total electric gross generation in Germany. Last but not least, I wanted to show a comparison between the electric gross generation by different renewable energy sectors. 

To choose the correct graphical forms, it is essential to first determine what information we want to convey. As mentioned in the previous paragraph, the most important message that I wanted to communicate to my audience is how the electric gross generation by renewable energy courses has changed in Germany since 1990 both in terms of total amount and in terms of composition. Stacked area plot does a perfect job at communicating these two pieces of information at the same time. Therefore, I chose stacked area plot as the major graphical form of this visualization. However, while stacked area plot does a good job communicating the change both in total amount and in composition, it still inevitably leaves the audiences with some major questions. For example, because the first subplot doesn't tell us how the total electric gross generation in Germany has changed over the years, the audience will have no idea how does the growth rate of electric gross generation by renewable energies compare to the total electric gross generation. In a hypothetical case, an even faster growth rate of the total electric gross generation in Germany would mean that renewable energies has become less important as sources of electric generation over the years. Thus, I also wanted to show the electric gross generation by renewable energy sources as a percentage of the total electric gross generation in Germany, as well as its longitudional change. 100% stacked area plot does a good job at representing longitudinal changes in percentages and therefore was chosen as the graphical form of my second subplot. Moreover, because it is somewhat difficult for the audience to draw a horizontal comparison between the electric gross generation by different forms of renewable energy, I added a line chart as my third subplot because it is well suited for the purpose.

Because there are three subplots in my visualization, it is thus important to organize them in a way that makes sense from both aesthetic and functional perspectives. First, because the most important messages are communicated through the first subplot, I decided to give the first subplot most importance by letting it occupy half of the canvas size and putting it on the left side so that the stacked area plot will immediate attract the audience's attention. On the other hand, the other two subplots occupy the right side of the canvas to make the visualization balanaced. In terms of color, I carefully chose some color keys from the seaborn library so that people can immediately and intuitively relate these colors to the names of the renewable energy sectors. For example, it is quite intuitive for people to associate the warm color of "amber" to the word "solar power", and the deep blue color to "hydroelectricity". Moreover, I also made sure that the color keys are consistent in subplot 1 and subplot 3 to eliminate any possibility of confussion. This deliberate color selection makes the visualization easier to read overall and enables the audience to extract information more quickly. Moreover, instead of using the default left position of the y-axis labels ticks and markers, I put the y-axis ticks and markers to the right side of the graphs for subplot 2 and subplot 3 to increase the balance of the visualization and to make it easier for people to read the text labels.
